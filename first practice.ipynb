```python
def reverseword(a):
    w=a.split(' ')
    n=[i[::-1] for i in w]
    ns=' '.join(n)
    return ns
a=input()
print(reverseword(a))
```

```python
import re
regex=r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
def check(email):
    if (re.fullmatch(regex,email)):
        print("YES")
    else:
        print("NO")
if __name__ == '__main__':
    email=input()
    check(email)
```

```python
import re
regex=r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b'
def check(list):
    for i in list:
        if i == re.findall(regex,list):
            print(i)
        else:
            print("0")
if __name__=='__main__':
    list=input().join(',')
    check(list)
```
